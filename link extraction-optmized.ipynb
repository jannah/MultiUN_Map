```python
from NetworkFinder import *
from Modules import multi_un_module as mun
import os, sys
from lxml import etree
from progressbar import AnimatedMarker, Bar, BouncingBar,\
                            Counter, ETA, Percentage, ProgressBar, SimpleProgress, FileTransferSpeed
print mun.PATH_TO_XML_FILES
```

```python
dirs = [dir for dir in os.listdir(mun.PATH_TO_XML_FILES) if dir not in['test', 'ENERGY','TOP_DOCS']]
files = {}
raw_filenames =[]
raw_full_filenames=[]
year_filenames = []
for dir in dirs:
    filenames = os.listdir(os.path.join(mun.PATH_TO_XML_FILES, dir))
    for filename in filenames:
        year_filenames.append(filename.replace('-en.xml', '')+'_'+dir)
        raw_filenames.append(filename)
        raw_full_filenames.append((dir, filename))
print len(year_filenames)
year_filenames = sorted(year_filenames)
# print year_filenames[-10:]
# print raw_full_filenames[-10
```

```python
import re
link_pattern = '([A-Z0-9._-]+/)+([A-z0-9._-]+)*'
doc_links = {}
documents = {}
doc_ids = {}
doc_paths = {}
index = 0
pbar = ProgressBar(widgets=[SimpleProgress(), Percentage(), Bar(), ETA()], maxval=len(raw_full_filenames)).start()
for (path, fn) in raw_full_filenames:
#     f = files[i]
    filename = os.path.join(mun.PATH_TO_XML_FILES,path, fn)
#     print filename
    documents[filename]={'attributes':[], 'content':[]}
    
    #with open(filename, 'r') as f:
    tree = etree.parse(filename)
    root = tree.getroot()
    documents[filename]['attributes'] = dict(root.attrib)
#     print documents[filename]['attributes']
    doc_id = documents[filename]['attributes']['id']
    doc_ids[doc_id]= documents[filename]['attributes']['n']
    doc_paths[doc_id] = filename
    xparas =  root.getchildren()[0].getchildren()[0].getchildren()
    content = []
    for xpara in xparas:
       content.append([mun.fix_unicode(sent.text.strip()) for sent in xpara if len(sent.text.strip())>0])
    text = "\n".join([ " ".join(c) for c in content])
    links = re.finditer(link_pattern, text)
    documents[filename]['links']=[]
    for link in links:
        l = link.group(0)
        documents[filename]['links'].append(l)
        if l not in doc_links:
            doc_links[l]=[]
        doc_links[l].append(filename)
    index+=1
    pbar.update(index)
pbar.finish()
   
# doc_ids
```

```python
link_list = sorted([ (l, len(doc_links[l]), len(set(doc_links[l])), doc_ids[l] if l in doc_ids else False, doc_paths[l] if l in doc_paths else False) 
                    for l in doc_links], key=lambda (key,val, val2, n, p): val, reverse= True)
with open('link_list.csv', 'w') as f:
    f.write('total_count, uniqu_count,id, n, fn, path\n')
    f.write("\n".join(['%d,%d,"%s","%s","%s","%s"'%(v,v2,k,l,p if p==False else p.split('\\')[-1], p) for (k,v,v2,l,p) in link_list]))
```
