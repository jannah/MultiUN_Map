```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
import csv

def draw_graph(graph):

    # extract nodes from graph
    nodes = set([n1 for n1, n2,w in graph] + [n2 for n1, n2,w in graph])

    # create networkx graph
    G=nx.Graph()

    # add nodes
    for node in nodes:
        G.add_node(node)

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # draw graph
    pos = nx.shell_layout(G)
    nx.draw(G, pos)

    # show graph
    plt.show()

# draw example
graph = []
with open('data/source_target.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile)
    graph = [(row[0], row[1], row[2]) for row in reader][1:]
#     for row in reader:
#         print row
        

# draw_graph(graph)
```
