```python
import shutil, json, os, simplejson, codecs
ROOT = 'C:\\Users\\Hassan\\Documents\\iSchool\\NLP\\United Nations\\'

doc_map_file = '../data/docmap.json'
doc_map = simplejson.loads(codecs.open(doc_map_file,'r', 'utf-8').read())
doc_map
```

```python
ids = [ (str(doc), str(doc_map[doc]['attributes']['id'])) for doc in doc_map]
links = [(str(doc_map[doc]['attributes']['id']), str(l['key']), l['value']) for doc in doc_map for l in doc_map[doc]['links']]
# links_out   = [v for v in [sum([1 for l in doc_map[doc]['links'] if l in doc_links]) for doc in doc_map] if v>0]

stats_in = {}
stats_out = {}
for (i,o,v) in links:
    if i not in stats_in:
        stats_in[i] = [0,0]
    if o not in stats_out:
        stats_out[o] = [0,0]
    stats_in[i][0]+=1
    stats_in[i][1]+=v
    stats_out[o][0]+=1
    stats_out[o][1]+=v
```

```python
links_in = sorted([(d, stats_in[d][0], stats_in[d][1]) for d in stats_in], key= lambda (d,s,t): t, reverse=True)
links_out =  sorted([(d, stats_out[d][0], stats_out[d][1]) for d in stats_out], key= lambda (d,s,t): t, reverse=True)
print links_in[:10]
print links_out[:10]
```

```python
top_1000 = set([d for (d,c,t) in links_in][:500]+[d for (d,c,t) in links_out][:500])
top_100 = set([d for (d,c,t) in links_in][:50]+[d for (d,c,t) in links_out][:50])
top_1000_paths = [doc_map[doc]['attributes']['path'] for doc in doc_map if doc_map[doc]['attributes']['id'] in top_1000]
top_100_paths = [doc_map[doc]['attributes']['path'] for doc in doc_map if doc_map[doc]['attributes']['id'] in top_100]
```

```python
def copy_files(folder, files):
    path ='..\\data\\%s'%folder
    if os.path.exists(path):
        shutil.rmtree(path)
    os.mkdir(path)
    for f in files:
        fn = f.split('\\')[-1]
        shutil.copy(f, os.path.join(path,fn))
copy_files('TOP_100', top_100_paths)
copy_files('TOP_1000', top_1000_paths)
```
