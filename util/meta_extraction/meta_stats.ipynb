```python
import json, zipfile
docmap_zip = zipfile.ZipFile('../docmap.zip')
scrape_zip = zipfile.ZipFile('../scrape/completed_scrape.zip')
# file = zip.read('text1.txt')
docmap = json.loads(docmap_zip.read('docmap.json','rb'))
scrape = json.load(docmap_zip.read('completed_scrape.json','rb'))
scrape_ns = dict([(str(scrape[i][1]),i) for i in range(len(scrape))])
```

```python
subjects = []
no_subjects = []
for doc in docmap:
    n = docmap[doc]['attributes']['n']
    if n in scrape_ns:
        i = scrape_ns[n]
        item = scrape[i]
        docmap[doc]['scrape'] = item[3]
#         print type(item[3])
        if isinstance(item[3], dict):
            if'Subjects' in item[3]:
                subjects+=docmap[doc]['scrape']['Subjects']
            else:
                no_subjects.append(doc)
#                 print doc, 'has no subjects'
        else:
            no_subjects.append(doc)
#             print doc, item[3]
    else:
        docmap[doc]['scrape'] = {}
print 'no subjects:', len(no_subjects)
print 'total subjects=', len(subjects), len(set(subjects))
# subjects = [docmap[doc]['scrape']['Subjects'] for doc in docmap if 'Subjects' in docmap[doc]['scrape']]
```
